```python
ratings = pd.read_csv('BX-Book-Ratings.csv',sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user-id','Book-Rating':'rating'},inplace=True)
```

```python
y=x[x].index
```

```python
ratings = ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user-id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
#convert bookpivot into sparse matrix
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
distances,suggestions = model.kneighbors(book_pivot.iloc[393, :].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1),n_neighbors=6)
    suggestions = suggestions.flatten().tolist()
    for i in range(1,len(suggestions)):
            print(book_pivot.index[suggestions[i]])   
```

```python
recommend_book('Animal Farm')
```
